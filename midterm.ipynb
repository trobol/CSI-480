```python
import pandas as pd
import sklearn as sklearn
from scipy import stats



data = pd.read_csv("/content/irisdata.csv")

data['sepal_width'] = data['sepal_width'].str.replace('%','').astype('int')

data['sepal_length'] = data['sepal_length'].str.replace('$','').astype('float')

data['petal_width'] = data['petal_width'].str.replace('t','').astype('float')


data[' petal_length '] = data[' petal_length '].str.replace(',', '').astype(float)

data['species'] = data['species'].astype(str)


data.set_index(data['ID'], inplace=True)

data
```
